```
from dataclasses import dataclass, asdict
from datetime import datetime


class config_utils:
    @dataclass
    class FabricConfig:
        fabric_environment: str
        config_workspace_id: str
        config_lakehouse_id: str
        edw_workspace_id: str
        edw_warehouse_id: str
        edw_warehouse_name: str
        edw_lakehouse_id: str
        edw_lakehouse_name: str
        legacy_synapse_connection_name: str
        synapse_export_shortcut_path_in_onelake: str
        full_reset: bool
        update_date: datetime = datetime.now()

        def get_attribute(self, attr_name: str) -> any:
            """Get attribute value by string name with error handling."""
            if hasattr(self, attr_name):
                return getattr(self, attr_name)
            else:
                raise AttributeError(f"FabricConfig has no attribute '{attr_name}'")

    def __init__(self, lakehouse: lakehouse_utils):
        self.lakehouse = lakehouse
        self.fabric_environments_table_uri = (
            f"{lakehouse.lakehouse_tables_uri()}config_fabric_environments"
        )

    @staticmethod
    def config_schema():
        return pa.schema(
            [
                ("fabric_environment", pa.string()),
                ("config_workspace_id", pa.string()),
                ("config_lakehouse_id", pa.string()),
                ("edw_workspace_id", pa.string()),
                ("edw_warehouse_id", pa.string()),
                ("edw_warehouse_name", pa.string()),
                ("edw_lakehouse_id", pa.string()),
                ("edw_lakehouse_name", pa.string()),
                ("legacy_synapse_connection_name", pa.string()),
                ("synapse_export_shortcut_path_in_onelake", pa.string()),
                ("full_reset", pa.bool_()),
                ("update_date", pa.timestamp("ms")),
            ]
        )

    def get_configs_as_dict(self, fabric_environment: str):
        table = (
            DeltaTable(self.fabric_environments_table_uri)
            .to_pyarrow_table()
            .to_pandas()
        )
        row = table[table["fabric_environment"] == fabric_environment]
        return row.iloc[0].to_dict() if not row.empty else None

    def get_configs_as_object(self, fabric_environment: str):
        row_dict = self.get_configs_as_dict(fabric_environment)
        return config_utils.FabricConfig(**row_dict) if row_dict else None
```

```
async def trigger_pipeline(client, workspace_id, pipeline_id, payload):
    trigger_url = f"/v1/workspaces/{workspace_id}/items/{pipeline_id}/jobs/instances?jobType=Pipeline"
    response = client.post(trigger_url, json=payload)
    if response.status_code != 202:
        raise Exception(
            f"❌ Failed to trigger pipeline: {response.status_code}\n{response.text}"
        )

    response_location = response.headers["Location"]
    # get last part of url from response_location
    job_id = response_location.rstrip("/").split("/")[-1]
    print(f"✅ Pipeline triggered. Pipeline: {job_id}")
    return job_id


async def trigger_notebook(client, workspace_id, notebook_id, payload):
    """
    Trigger a Fabric notebook job via REST.

    :param client:        an authenticated HTTP client, e.g. requests.Session()
    :param workspace_id:  your Fabric workspace GUID
    :param notebook_id:   the item ID of the notebook to run (from the URL in the UI)
    :param payload:       dict of notebook parameters
    :returns:             the launched job ID
    """
    # Note jobType=Notebook here
    trigger_url = (
        f"/v1/workspaces/{workspace_id}"
        f"/items/{notebook_id}"
        "/jobs/instances?jobType=Notebook"
    )

    response = client.post(trigger_url, json=payload)
    if response.status_code != 202:
        raise Exception(
            f"❌ Failed to trigger notebook: {response.status_code}\n{response.text}"
        )

    # Location header comes back as “…/jobs/instances/{jobId}”
    location = response.headers.get("Location", "")
    job_id = location.rstrip("/").split("/")[-1]

    print(f"✅ Notebook triggered. Job ID: {job_id}")
    return job_id


async def check_notebook(
    client, progress: Progress, task, workspace_id: str, notebook_id: str, job_id: str
):
    """
    Poll a Fabric notebook job for its current status.

    :param client:        an authenticated HTTP client (e.g. requests.Session())
    :param progress:      a rich.Progress instance
    :param task:          the task ID returned from progress.add_task()
    :param workspace_id:  your Fabric workspace GUID
    :param notebook_id:   the item ID of the notebook you launched
    :param job_id:        the job instance ID you received when triggering
    :returns:             the notebook’s current status
    """
    # NOTE: using the notebook_id here instead of a pipeline_id
    status_url = (
        f"/v1/workspaces/{workspace_id}/items/{notebook_id}/jobs/instances/{job_id}"
    )
    terminal_states = {"Completed", "Failed", "Cancelled", "Deduped"}

    resp = client.get(status_url)
    resp.raise_for_status()
    status = resp.json().get("status")

    # update the progress display with an appropriate emoji
    if status == "Completed":
        progress.update(task, description=f"Notebook {job_id}: ✅ {status}")
    elif status == "Failed":
        progress.update(task, description=f"Notebook {job_id}: ❌ {status}")
    else:
        progress.update(task, description=f"Notebook {job_id}: ⚠️ {status}")

    return status


async def check_pipeline(
    client, progress: Progress, task, workspace_id, pipeline_id, job_id
):
    status_url = (
        f"/v1/workspaces/{workspace_id}/items/{pipeline_id}/jobs/instances/{job_id}"
    )
    terminal_states = {"Completed", "Failed", "Cancelled", "Deduped"}

    response = client.get(status_url)
    data = response.json()
    status = data.get("status")

    if status == "Failed" and "failureReason" in data:
        fr = data["failureReason"]
        msg = fr.get("message", "")
        if fr.get("errorCode") == "RequestExecutionFailed" and "NotFound" in msg:
            print(f"[INFO] Transient check-failure, retrying: {fr.get('message')!r}")
            return None
        return status, msg

    print(f"[STATUS] Pipeline job_id={job_id} current status: {status}")

    return status, ""
```

```
from deltalake import write_deltalake
import pyarrow as pa
import pandas as pd
from datetime import datetime
import numpy as np


class synpase_extract_utils:
    def __init__(
        self,
        datasource_name: str,
        datasource_location: str,
        target_workspace_id,
        target_lakehouse_id,
    ):
        self.datasource_name = datasource_name
        self.datasource_location = datasource_location
        self.log_table_uri = f"abfss://{target_workspace_id}@onelake.dfs.fabric.microsoft.com/{target_lakehouse_id}/Tables/log_synapse_extracts"
        self.config_table_uri = f"abfss://{target_workspace_id}@onelake.dfs.fabric.microsoft.com/{target_lakehouse_id}/Tables/config_synapse_extracts"

    def get_extract_sql_template(self):
        return f"""

    IF NOT EXISTS (
        SELECT
            *
        FROM
            sys.external_file_formats 
        WHERE
            name = 'parquet'
    )
    CREATE EXTERNAL FILE FORMAT [parquet] WITH (FORMAT_TYPE = PARQUET, DATA_COMPRESSION = 'org.apache.hadoop.io.compress.SnappyCodec')

    IF NOT EXISTS (
        SELECT
            *
        FROM
            sys.external_data_sources
        WHERE
            name = '{self.datasource_name}'
    ) 
    CREATE EXTERNAL DATA SOURCE [{self.datasource_name}] WITH (
        LOCATION = '{self.datasource_location}'
    ) 

    IF OBJECT_ID('exports.@TableSchema_@TableName', 'U') IS NOT NULL
    DROP EXTERNAL TABLE exports.@TableSchema_@TableName

    CREATE EXTERNAL TABLE exports.@TableSchema_@TableName WITH (
        LOCATION = 'exports/@TableSchema_@TableName', 
        DATA_SOURCE = [{self.datasource_name}],
        FILE_FORMAT = parquet 
    ) AS
    SELECT
        *
    FROM
        [@TableSchema].[@TableName]
    @PartitionClause

    """

    async def insert_log_record(
        self,
        execution_id,
        cfg_synapse_connection_name,
        source_schema_name,
        source_table_name,
        extract_file_name,
        partition_clause,
        status,
        error_messages,
        start_date=None,
        finish_date=None,
        update_date=None,
    ):
        # Use UTC timestamp in int64 format: YYYYMMDDHHMMSSmmm
        def now_int64():
            return np.int64(datetime.utcnow().strftime("%Y%m%d%H%M%S%f")[:-3])

        # Default the dates only if they are not provided
        start_ts = start_date if start_date is not None else now_int64()
        finish_ts = finish_date if finish_date is not None else now_int64()
        update_ts = update_date if update_date is not None else now_int64()

        # Step 1: Prepare schema
        schema = pa.schema(
            [
                ("execution_id", pa.string()),
                ("cfg_synapse_connection_name", pa.string()),
                ("source_schema_name", pa.string()),
                ("source_table_name", pa.string()),
                ("extract_file_name", pa.string()),
                ("partition_clause", pa.string()),
                ("status", pa.string()),
                ("error_messages", pa.string()),
                ("start_date", pa.int64()),
                ("finish_date", pa.int64()),
                ("update_date", pa.int64()),
            ]
        )

        # Step 2: Create DataFrame
        df = pd.DataFrame(
            [
                {
                    "execution_id": execution_id,
                    "cfg_synapse_connection_name": cfg_synapse_connection_name,
                    "source_schema_name": source_schema_name,
                    "source_table_name": source_table_name,
                    "extract_file_name": extract_file_name,
                    "partition_clause": partition_clause,
                    "status": status,
                    "error_messages": error_messages,
                    "start_date": start_ts,
                    "finish_date": finish_ts,
                    "update_date": update_ts,
                }
            ]
        )

        # Step 3: Convert to Arrow Table
        # Downcast to milliseconds to match delta-rs compatibility

        table = pa.Table.from_pandas(df, schema=schema, preserve_index=False)

        # Step 4: Write to Delta (append)
        write_deltalake(self.log_table_uri, table, mode="append", schema=schema)
```

```
def timestamp_now():
    return int(datetime.utcnow().strftime("%Y%m%d%H%M%S%f")[:-3])
```

```
def prepare_extract_payloads(
    tables: List[ExtractConfigDetails],
    sql_template: str,
    run_year: int,
    run_month: int,
    run_day: int,
) -> None:
    for table in tables:
        script = (
            sql_template.replace("@TableSchema", table.source_schema_name)
            .replace("@TableName", table.source_table_name)
            .replace("@PartitionClause", table.partition_clause)
            .replace("@year", str(run_year))
            .replace("@month", str(run_month))
            .replace("@day", str(run_day))
        )

        table.data_pipeline_payload = {
            "executionData": {
                "parameters": {
                    "script_content": script,
                    "table_schema_and_name": f"{table.source_schema_name}_{table.source_table_name}",
                }
            }
        }
```

```
async def trigger_chunked(tables: List[ExtractConfigDetails], max_parallel: int):
    for i in range(0, len(tables), max_parallel):
        chunk = tables[i : i + max_parallel]
        job_ids = await asyncio.gather(
            *[
                trigger_pipeline(
                    client=client,
                    workspace_id=workspace_id,
                    pipeline_id=pipeline_id,
                    payload=table.data_pipeline_payload,
                )
                for table in chunk
            ]
        )

        # assign attributes back to each table object
        for table, job_id in zip(chunk, job_ids):
            table.job = DataPipelineJob(job_id=job_id, status="NotStarted")
            table.execution_id = table.execution_id or str(uuid.uuid4())
            table.start_date = timestamp_now()
```

```
# get sql template, trigger pipelines in batches

sql_template = su.get_extract_sql_template()
prepare_extract_payloads(
    tables=tables_to_extract,
    sql_template=sql_template,
    run_year=run_year,
    run_month=run_month,
    run_day=run_day,
)

await trigger_chunked(tables=tables_to_extract, max_parallel=1)
```

```
poll_interval = 10  # seconds
terminal_states = {"Completed", "Failed", "Cancelled", "Deduped"}
conn = notebookutils.data.connect_to_artifact("WH")

log_tasks = []
with Progress(transient=False) as progress:
    # create progress task per table
    for table in tables_to_extract:
        table.progress_task = progress.add_task(
            f"{table.source_schema_name}_{table.source_table_name}", total=1
        )

    while True:
        any_active = False

        for table in tables_to_extract:
            full_name = f"{table.source_schema_name}_{table.source_table_name}"
            prev_status = table.job.status

            try:
                status, error_message = await check_pipeline(
                    client=client,
                    progress=progress,
                    task=table.progress_task,
                    workspace_id=workspace_id,
                    pipeline_id=pipeline_id,
                    job_id=table.job.job_id,
                )
            except Exception as e:
                ts = timestamp_now()
                log_tasks.append(
                    su.insert_log_record(
                        execution_id=str(table.execution_id),
                        cfg_synapse_connection_name=table.get_attribute(
                            "synapse_connection_name"
                        ),
                        source_schema_name=table.source_schema_name,
                        source_table_name=table.source_table_name,
                        extract_file_name=full_name,
                        partition_clause=table.partition_clause.replace(
                            "@year", str(run_year)
                        )
                        .replace("@month", str(run_month))
                        .replace("@day", str(run_day)),
                        status="Failed",
                        error_messages=str(e),
                        start_date=table.start_date,
                        finish_date=ts,
                        update_date=ts,
                    )
                )
                table.job.status = "Failed"
                continue

            if status is None:
                any_active = True
                progress.update(
                    table.progress_task, description=f"Job {full_name}: retrying..."
                )
                ts = timestamp_now()
                log_tasks.append(
                    su.insert_log_record(
                        execution_id=str(table.execution_id),
                        cfg_synapse_connection_name=table.get_attribute(
                            "synapse_connection_name"
                        ),
                        source_schema_name=table.source_schema_name,
                        source_table_name=table.source_table_name,
                        extract_file_name=full_name,
                        partition_clause=table.partition_clause.replace(
                            "@year", str(run_year)
                        )
                        .replace("@month", str(run_month))
                        .replace("@day", str(run_day)),
                        status="Retrying",
                        error_messages=error_message,
                        start_date=table.start_date,
                        finish_date=None,
                        update_date=ts,
                    )
                )
                continue

            table.job.status = status
            ts = timestamp_now()

            if status != prev_status or status in terminal_states:
                log_tasks.append(
                    su.insert_log_record(
                        execution_id=str(table.execution_id),
                        cfg_synapse_connection_name=table.get_attribute(
                            "synapse_connection_name"
                        ),
                        source_schema_name=table.source_schema_name,
                        source_table_name=table.source_table_name,
                        extract_file_name=full_name,
                        partition_clause=table.partition_clause.replace(
                            "@year", str(run_year)
                        )
                        .replace("@month", str(run_month))
                        .replace("@day", str(run_day)),
                        status=status,
                        error_messages=error_message,
                        start_date=table.start_date,
                        finish_date=ts if status in terminal_states else None,
                        update_date=ts,
                    )
                )

            if status not in terminal_states:
                any_active = True
                progress.update(
                    table.progress_task, description=f"Job {full_name}: {status}"
                )
            else:
                progress.update(
                    table.progress_task,
                    description=f"Job {full_name}: {status}",
                    advance=1,
                )
                progress.refresh()

            # synapse_connection_name = configs.get_attribute(table.cfg_legacy_synapse_connection_name)

        if not any_active:
            break

        await asyncio.sleep(poll_interval)

    summary = Table(title="Pipeline Jobs Status Summary")
    summary.add_column("Job ID", style="bold")
    summary.add_column("Table Name", style="bold")
    summary.add_column("Status", style="bold")
    for table in tables_to_extract:
        summary.add_row(
            str(table.job.job_id), table.source_table_name, table.job.status
        )
    progress.console.print(summary)

await asyncio.gather(*log_tasks)
```
